```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data.csv')
df
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = pd.value_counts(df.race)
race_count
```

```python
# What is the average age of men?
df.age[df.sex=='Male'].mean()
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = df.education[df.education=='Bachelors'].count()*100/df.education.count()
percentage_bachelors
```

```python
df50k = df[df.salary=='>50K']
df50k
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?
higher_education = df.education[(df.education=='Bachelors') | (df.education=='Masters') | (df.education=='Doctorate')].count()
education_counts = pd.value_counts(df.education)
education_counts
```

```python
lower_education = df.education.count() - higher_education
lower_education
```

```python
higher_education_rich = df50k.education[(df50k.education=='Bachelors') | (df50k.education=='Masters') | (df50k.education=='Doctorate')].count()
lower_education_rich = df50k.education.count() - higher_education_rich
lower_education_rich
```

```python
higher_education_rich = higher_education_rich*100/df.education.count()
lower_education_rich = lower_education_rich*100/df.education.count()
higher_education_rich
```
