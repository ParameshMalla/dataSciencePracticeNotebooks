### Random Variables

Two independent dices are rolled

```
d={(i,j):i+j for i in range(1,7) for j in range(1,7)}
d
```

```
from collections import defaultdict
dinv = defaultdict(list)
for i,j in d.items():
  dinv[j].append(i)
dinv
```

dinv contains all the possible sums of two dices

```
dinv[11]
```

Probability mass function for sum of two dices

```
X={i:len(j)/36. for i,j in dinv.items()}
X
```

The probability that half the product of three dice will exceed the their sum

```
d={(i,j,k):((i*j*k)/2>i+j+k) for i in range(1,7)
                               for j in range(1,7)
                                 for k in range(1,7)}
dinv = defaultdict(list)
for i,j in d.items():
  dinv[j].append(j)
dinv
```

```
X={i:len(j)/6.0**3 for i,j in dinv.items()}
X
```
