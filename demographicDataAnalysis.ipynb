```python
import pandas as pd
```

```python
df = pd.read_csv('/content/adult.data.csv')
df
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = pd.value_counts(df.race)
race_count
```

```python
# What is the average age of men?
df.age[df.sex=='Male'].mean()
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = df.education[df.education=='Bachelors'].count()*100/df.education.count()
percentage_bachelors
```

```python
df50k = df[df.salary=='>50K']
df50k
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?
higher_education = df.education[(df.education=='Bachelors') | (df.education=='Masters') | (df.education=='Doctorate')].count()
education_counts = pd.value_counts(df.education)
education_counts
```

```python
lower_education = df.education.count() - higher_education
lower_education
```

```python
higher_education_rich = df50k.education[(df50k.education=='Bachelors') | (df50k.education=='Masters') | (df50k.education=='Doctorate')].count()
lower_education_rich = df50k.education.count() - higher_education_rich
lower_education_rich
```

```python
higher_education_rich = higher_education_rich*100/df.education.count()
lower_education_rich = lower_education_rich*100/df.education.count()
higher_education_rich
```

```python
df['hours-per-week'].describe()
```

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
(df['hours-per-week'] == min_work_hours).values.sum()
```

```python
(df50k['hours-per-week'] == min_work_hours).values.sum()
```

```python
rich_percentage = (df50k['hours-per-week'] == min_work_hours).values.sum()*100/(df['hours-per-week'] == min_work_hours).values.sum()
rich_percentage
```

```python
# What country has the highest percentage of people that earn >50K?
df50k['native-country'].value_counts().rename_axis('native-country').reset_index(name='>50k')
```

```python
df['native-country'].value_counts().rename_axis('native-country').reset_index(name='all')
```

```python
perdf = pd.merge(df50k['native-country'].value_counts().rename_axis('native-country').reset_index(name='>50k'),df['native-country'].value_counts().rename_axis('native-country').reset_index(name='all'),how='outer',on='native-country')
perdf
```

```python
perdf['percentage'] = perdf['>50k']*100/perdf['all']
perdf = perdf.sort_values('percentage', ascending=False)
perdf
```

```python
perdf.loc[perdf.index[0],'native-country']
```

```python
# Identify the most popular occupation for those who earn >50K in India.
(df50k[df50k['native-country'] == 'India'].occupation.value_counts()).index[0]
```

```python
countries = pd.DataFrame(df['native-country'].value_counts().reset_index())
countries.columns = ['country','count']
countries50k = pd.DataFrame(df50k['native-country'].value_counts().reset_index())
countries50k.columns = ['country','count50k']
c_df = pd.merge(countries, countries50k, how='outer',on='country')
c_df
```
